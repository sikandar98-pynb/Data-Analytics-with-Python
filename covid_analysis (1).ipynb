```python
#finding out the number of active cases
world_covid_cases['Active'] = world_covid_cases.Confirmed - world_covid_cases.Deaths - world_covid_cases.Recovered
```

```python
#checking the number of active cases
world_covid_cases.head(5)
```

```python
#droping unwanted columns
world_covid_cases = world_covid_cases.drop(['SNo', 'Last Update'],axis =1)
```

```python
#verify the new table
world_covid_cases.head(5)
```

```python
#creating a dictionary containing the new column headers and passing it through the dataframe
dict_1 = {'Province/State':'State','Country/Region': 'Country','ObservationDate':'Date'
         }
```

```python
world_covid_cases.rename(columns = dict_1, inplace = True)
world_covid_cases.head(5)
```

```python
#show the distint values in the column country so that if required, renaming can be done
world_covid_cases.Country.unique()
```

```python
#extracting the data from the date-time column using the delimeter 'space'
world_covid_cases.Date = (world_covid_cases.Date.str.split( ).str[0])
world_covid_cases.head(5)
```

```python
world_covid_cases['Active%'] = (world_covid_cases.Active)/(world_covid_cases.Confirmed)*100
```

```python
world_covid_cases
```

```python
active_cases = world_covid_cases.groupby('Country')['Confirmed', 'Deaths', 'Recovered', 'Active'].sum().reset_index()
```

```python
active_cases_analysis = active_cases.sort_values(by='Active', ascending=False)
active_cases_analysis = active_cases_analysis.reset_index(drop=True)
active_cases_analysis.style.background_gradient(cmap='Reds')
```

```python
world_covid_cases['Year'] = world_covid_cases['Date'].str[-4:]
```

```python
world_covid_cases
```

```python
top_10_countries = active_cases_analysis.head(10)
```

```python
top_10_countries.T
```

```python
plt.plot(top_10_countries['Country'],top_10_countries['Active'])
plt.title('Active cases per Year')
plt.xlabel('Country')
plt.ylabel('Cases')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(10)
plt.plot(top_10_countries['Country'],top_10_countries['Active'],color='red',marker='o')
plt.show()
```
